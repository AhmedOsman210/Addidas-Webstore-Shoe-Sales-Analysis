```python
# Load Excel files
df_country = pd.read_csv("country_dim.csv")
df_shoes = pd.read_csv("shoes_dim.csv")
df_sales = pd.read_csv("shoes_fact.csv")


# Display first few rows of each dataset
df_country.head(), df_shoes.head(), df_sales.head()
```

```python
# Get basic info about each dataset
df_country.info()
df_shoes.info()
df_sales.info()

# Check for missing values in each dataset
print(df_country.isnull().sum())
print(df_shoes.isnull().sum())
print(df_sales.isnull().sum())
```

```python
# Fill missing price values with the mean in df_sales
df_sales['price'] = df_sales['price'].fillna(df_sales['price'].mean())
```

```python
# Merge the datasets
df_merged = df_sales.merge(df_shoes, on="id", how="left")  # Merge sales with shoes data
df_merged = df_merged.merge(df_country, on="country_code", how="left")  # Merge with country data

# Display the first few rows of the merged dataset
df_merged.head()
```

```python
# Basic overview of the merged dataset
df_merged.info()

# Descriptive statistics for numerical columns
df_merged.describe()

# Check for missing values again
df_merged.isnull().sum()
```

```python
# Distribution of shoe prices
plt.figure(figsize=(10, 5))
sns.histplot(df_merged['price'], bins=30, kde=True, color="blue")
plt.title("Price Distribution of Adidas Shoes")
plt.xlabel("Price (€)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Shoe availability by category
plt.figure(figsize=(12, 6))
sns.countplot(data=df_merged, y="category", order=df_merged["category"].value_counts().index, palette="viridis")
plt.title("Shoe Availability by Category")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

```python
# Total shoe availability by country
sales_by_country = df_merged.groupby('country_code')['availability'].sum().sort_values()

plt.figure(figsize=(10, 5))
sales_by_country.plot(kind="barh", color="green")
plt.title("Total Shoe Availability by Country")
plt.xlabel("Total Availability")
plt.ylabel("Country Code")
plt.show()
```

```python
# Average price by gender
plt.figure(figsize=(8, 5))
sns.boxplot(data=df_merged, x='gender', y='price', palette="coolwarm")
plt.title("Average Price by Gender")
plt.xlabel("Gender")
plt.ylabel("Price (€)")
plt.show()

# Average price by category
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_merged, x='category', y='price', palette="coolwarm")
plt.title("Average Price by Category")
plt.xlabel("Category")
plt.ylabel("Price (€)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert the 'date' column to datetime (handling various formats)
df_merged['date'] = pd.to_datetime(df_merged['date'], errors='coerce')

# Group by date and get the total sales or availability
sales_over_time = df_merged.groupby('date')['availability'].sum()

# Plot the trend over time
plt.figure(figsize=(10, 5))
sales_over_time.plot(kind="line", color="blue")
plt.title("Total Shoe Availability Over Time")
plt.xlabel("Date")
plt.ylabel("Total Availability")
plt.xticks(rotation=45)
plt.show()
```

```python
# Select only numerical columns
numerical_df = df_merged.select_dtypes(include=['float64', 'int64'])

# Check correlation between numerical features
correlation_matrix = numerical_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select numerical columns for clustering (e.g., price, availability)
df_clustering = df_merged[['price', 'availability']].dropna()  # Remove rows with NaN

# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clustering)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3)  # Let's try 3 clusters
df_merged['cluster'] = kmeans.fit_predict(df_scaled)

# Visualize clusters
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_merged, x='price', y='availability', hue='cluster', palette='viridis')
plt.title("K-Means Clustering of Shoes")
plt.xlabel("Price")
plt.ylabel("Availability")
plt.legend()
plt.show()
```

```python
# Group by country code and calculate total sales per country
sales_per_country = df_merged.groupby('country_code')['availability'].sum().reset_index()

# Sort by total sales
sales_per_country_sorted = sales_per_country.sort_values(by='availability', ascending=False)

# Display and plot the results
plt.figure(figsize=(12,6))
plt.bar(sales_per_country_sorted['country_code'], sales_per_country_sorted['availability'])
plt.title('Sales by Country')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
